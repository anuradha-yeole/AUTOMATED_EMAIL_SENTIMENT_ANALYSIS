```python
def calculate_body_type(row):

    waist=row['Waist']
    bust=row['Bust']
    hip=row['Hip']
    waist_to_hip_ratio = waist / hip
    bust_to_waist_ratio = bust / waist

    if bust_to_waist_ratio < 0.75 and waist_to_hip_ratio > 0.85:
        return "Pear"
    elif bust_to_waist_ratio > 0.75 and waist_to_hip_ratio < 0.75:
        return "Apple"
    elif bust_to_waist_ratio < 0.75 and 0.75 <= waist_to_hip_ratio <= 0.85:
        return "Hourglass"
    elif bust_to_waist_ratio > 0.75 and waist_to_hip_ratio > 0.85:
        return "Inverted Triangle"
    else:
        return "Rectangle"

# Apply the function to each row and create a new column "Body Type"
df['Body Type'] = df.apply(calculate_body_type, axis=1)
```

```python
import os
import json
import pandas as pd

# Custom body type calculator function
def calculate_body_type(row):
    waist = float(row.get('waist', 0))
    bust = float(row.get('bust', 0))
    hip = float(row.get('hips', 0))
    waist_to_hip_ratio = waist / hip
    bust_to_waist_ratio = bust / waist

    if bust_to_waist_ratio < 0.75 and waist_to_hip_ratio > 0.85:
        return "Pear"
    elif bust_to_waist_ratio > 0.75 and waist_to_hip_ratio < 0.75:
        return "Apple"
    elif bust_to_waist_ratio < 0.75 and 0.75 <= waist_to_hip_ratio <= 0.85:
        return "Hourglass"
    elif bust_to_waist_ratio > 0.75 and waist_to_hip_ratio > 0.85:
        return "Inverted Triangle"
    else:
        return "Rectangle"

# Initialize DataFrame outside the loop
df = pd.DataFrame()

garment_folder_path = "/content/VIBE/dataset/onepiece/garment_measurement"
meta_description_folder_path = "/content/VIBE/dataset/onepiece/meta_descriptions"
model_description_folder_path = "/content/VIBE/dataset/onepiece/model_measurement"
image_folder_path = "/content/VIBE/dataset/onepiece/img_url"

# Loop for garment data
for garment_filename in os.listdir(garment_folder_path):
    if garment_filename.endswith("_meta.txt"):
        # Construct the corresponding image URL filename
        image_filename = garment_filename.replace("_meta.txt", "_imgsrc.txt")

        garment_file_path = os.path.join(garment_folder_path, garment_filename)
        image_file_path = os.path.join(image_folder_path, image_filename)
        meta_description_file_path = os.path.join(meta_description_folder_path, garment_filename)

        try:
            # Read JSON data from the garment text file
            with open(garment_file_path, 'r') as garment_file:
                garment_json_data = json.load(garment_file)

            try:
                with open(image_file_path, 'r') as image_file:
                    image_url = image_file.read().strip()
            except FileNotFoundError:
                print(f"Image file not found for {garment_filename}. Skipping.")
                continue

            try:
                # Read content from the meta-description text file
                with open(meta_description_file_path, 'r') as meta_description_file:
                    description = meta_description_file.read().strip()
            except FileNotFoundError:
                print(f"Meta-description file not found for {garment_filename}. Skipping.")
                continue

            product_name = garment_filename.split("#")[0]
            # Dynamically identify the key containing the dimension information
            dimension_key = next((key for key, value in garment_json_data.items() if isinstance(value, list)), None)

            if dimension_key:
                # Extract the relevant information and append it to the DataFrame
                dimensions = garment_json_data[dimension_key]

                # Extract column names dynamically from dimension names
                column_names = [dimension['dimension_name'] for dimension in dimensions]

                # Extract values and create a dictionary with column names as keys
                values = {dimension['dimension_name']: dimension['value'] for dimension in dimensions}

                # Add the image_url, description, and product_name columns to the dictionary
                values['image_url'] = image_url
                values['description'] = description
                values['product_name'] = product_name

                # Append values to the DataFrame
                df = df.append(values, ignore_index=True)
            else:
                print(f"Skipping file {garment_filename} as it does not contain the expected dimension structure.")
        except json.JSONDecodeError:
            print(f"Error decoding JSON in file {garment_filename}.")

# Loop for model data
for model_filename in os.listdir(model_description_folder_path):
    if model_filename.endswith(".txt"):
        model_file_path = os.path.join(model_description_folder_path, model_filename)

        try:
            # Read JSON data from the model description file
            with open(model_file_path, 'r') as model_file:
                model_data = json.load(model_file)

            # Create lists to store multiple models' data
            model_names = []
            body_types = []

            # Iterate over each model name in the dictionary
            for model_name, model_values in model_data.items():
                # Calculate body type using the custom function
                body_type = calculate_body_type(model_values)

                # Append model name and body type to lists
                model_names.append(model_name)
                body_types.append(body_type)

            # If there are models, append the lists to the DataFrame
            if model_names:
                df = df.append({'model_name': model_names, 'body_type': body_types}, ignore_index=True)

        except json.JSONDecodeError:
            print(f"Error decoding JSON in file {model_filename}.")

# Display the resulting DataFrame
print(df)
```
